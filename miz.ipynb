```python
movies = movie_data.merge(movie_credits, on = 'title')
```

```python
# important data
# genres
# id
# keywords
# title
# release date(task)
# cast
# crew

movies = movies[['movie_id', 'overview', 'title', 'genres', 'keywords', 'title', 'cast', 'crew']]
```

```python
# preprocessing --> formatting data
movies.isnull().sum()
movies.dropna(inplace=True)

# movies.duplicated().sum()
```

```python
# getting genres for a item
movies.iloc[0].genres
```

```python
import ast
# ast.literal_eval() # convert string of list to list
```

```python
def extract(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(extract)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(extract)
```

```python
def extract_cast(obj):
    L = []
    c = 0
    for i in ast.literal_eval(obj):
        if c != 3:
            L.append(i['name'])
            c += 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(extract_cast)
```

```python
def extract_crew(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(extract_crew)
# movies['overview'].apply(lambda x:x.split())
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
# print((movies['overview']))
# print((movies['crew']))
```

```python
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df = movies[['movie_id', 'title', 'tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
# new_df['tags'] = movies[['tags']]
new_df.head()
# movies.head()
# movies.info()
# new_df.columns.duplicated()
# new_df['title']
# print(type(new_df))
# new_df = new_df.loc[:,~new_df.columns.duplicated()].copy()
# new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
#vectorization --> text to vector 
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names()
```

```python
# applying stemming
```

```python
import nltk #famous library for nlp
```

```python
# !pip install nltk

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stemer(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stemer)
```

```python
# will calculate the cosin distance (angle) not eucledian distance
# because eucledian distance in higher dimension is not a good measure
# cosine distance --> distance inversaly proporational to similarities 
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommender(movie):
    movie_ind = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_ind]
    movies = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    for i in movies:
        print (new_df.iloc[i[0]].title)
```
